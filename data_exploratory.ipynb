```python
import pandas as pd
import numpy as np
import pickle
from collections import Counter
```

```python
#Loading the mtg dataset genre file
mtg = pd.read_csv('/content/mtg_cleaned.csv', index_col = [0])
```

```python
mtg
```

```python
# Drop rows with NaN values in the 'genre' column
mtg_ = mtg.dropna(subset=['genre'])

# Split the genres and count occurrences
all_genres = [genre.split(', ') for genre in mtg_['genre']]
flat_genres = [genre for sublist in all_genres for genre in sublist]
genre_counts = Counter(flat_genres)

# Convert the Counter to a DataFrame for better visualization
genre_counts_df = pd.DataFrame(list(genre_counts.items()), columns=['genre', 'count']).sort_values(by='count', ascending=False, ignore_index = True)
genre_counts_df
```

```python
# Load the data from the pickle file
with open('/content/musiccaps_tags.pkl', 'rb') as file:
    musicap = pickle.load(file)
```

```python
musicap
```

```python
flat_musicap = [genre for sublist in musicap for genre in sublist]

# Count occurrences using Counter
genre_counts_cap = Counter(flat_musicap)

# Convert the Counter to a DataFrame for better visualization
genre_counts_df_cap = pd.DataFrame(list(genre_counts_cap.items()), columns=['genre', 'count'])

# Sort the DataFrame by count in descending order
genre_counts_df_cap = genre_counts_df_cap.sort_values(by='count', ascending=False, ignore_index = True)

# Display the result
genre_counts_df_cap
```

```python
genre_counts_df_cap['genre'] = genre_counts_df_cap['genre'].apply(lambda x: x.replace(' ', ''))
```

```python
#Creating a list of all common genres
genre_common = list(set(genre_counts_df['genre']).intersection(genre_counts_df_cap['genre']))
```

```python
genre_common
```

```python
#creating a list of all the genres only occuring in the mtg dataset
unique_mtg = list(set(genre_counts_df['genre']) - set(genre_counts_df_cap['genre']))

#creating a list of all the genres only occuring in the musiccaps dataset
unique_caps = list(set(genre_counts_df_cap['genre']) - set(genre_counts_df['genre']))
```

```python
unique_mtg
```

```python
unique_caps
```

```python
'''creating a dict that has the unique_mtg elements as key and the value as all the musiccaps elements that have the mtg element in them'''
common_dict = {}

for key in unique_mtg:
  common_dict[key] = [value for value in unique_caps if key in value]

print(common_dict)
```
