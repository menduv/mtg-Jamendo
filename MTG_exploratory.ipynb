```python
import re

# Example string
s = 'genre---atmospheric genre---classical genre---contemporary genre---soundtrack instrument---cello instrument---orchestra instrument---piano instrument---violin'

# Extract everything between --- and genre, instrument or mood
result_genre = re.findall(r'e---(.*?)(?: genr| instrument| mood|$)', s)
result_instrument = re.findall(r'ment---(.*?)(?: genre| instru| mood|$)', s)
result_mood = re.findall(r'theme---(.*?)(?: genre| instrument| mood|$)', s)

print(result_genre)
print(result_instrument)
print(result_mood)
```

```python
# Define a function to extract values based on the provided regex pattern
def extract_values(tag_string, category):
    pattern = f'{category}(.*?)(?: gen| instru| mood|$)'
    values = re.findall(pattern, tag_string)
    return ', '.join(values) if values else ''

# Apply the function to each row in the DataFrame
df['genre'] = df['tags'].apply(lambda x: extract_values(x, 're---'))
df['instrument'] = df['tags'].apply(lambda x: extract_values(x, 'ment---'))
df['mood'] = df['tags'].apply(lambda x: extract_values(x, 'theme---'))
```
